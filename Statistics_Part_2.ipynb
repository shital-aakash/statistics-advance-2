3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import numpy as np #code
def bernoulli_pdf(x, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    x: The value for which to calculate the probability. Must be either 0 or 1.
    p: The probability of success (the probability of x=1).

  Returns:
    The probability of x given the probability of success p.
  """

  if x == 0:
    return (1 - p)
  elif x == 1:
    return p
  else:
    return 0
p = 0.6
x = 1
probability = bernoulli_pdf(x, p)
print(f"Probability of x={x} with p={p}: {probability}")
x = 0
probability = bernoulli_pdf(x, p)
print(f"Probability of x={x} with p={p}: {probability}")
x = 2
probability = bernoulli_pdf(x, p)
print(f"Probability of x={x} with p={p}: {probability}")
```

6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np #code
import matplotlib.pyplot as plt
def discrete_uniform_cdf(low, high, x):
    """
    Calculates the CDF of a discrete uniform distribution.

    Args:
        low (int): Lower bound of the distribution (inclusive).
        high (int): Upper bound of the distribution (inclusive).
        x (array-like): Values at which to evaluate the CDF.

    Returns:
        numpy.ndarray: CDF values corresponding to x.
    """
    x = np.array(x)
    cdf = np.zeros_like(x, dtype=float)
    for i in range(len(x)):
        if x[i] < low:
            cdf[i] = 0
        elif x[i] >= high:
            cdf[i] = 1
        else:
            cdf[i] = (np.floor(x[i]) - low + 1) / (high - low + 1)
    return cdf
low = 1
high = 6
x = np.arange(low - 1, high + 2)
cdf = discrete_uniform_cdf(low, high, x)
plt.step(x, cdf, where='post')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.title(f'CDF of Discrete Uniform Distribution (Low={low}, High={high})')
plt.xticks(x)
plt.grid(True)
plt.show()
```

14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np #code
from scipy.stats import norm
def calculate_z_score(data_point, data_set):
    """
    Calculate the Z-score for a given data point within a data set.

    Args:
        data_point (float or int): The individual data point.
        data_set (list or numpy.ndarray): The data set.

    Returns:
        float: The calculated Z-score.
    """
    mean = np.mean(data_set)
    std_dev = np.std(data_set)
    z_score = (data_point - mean) / std_dev
    return z_score
def compare_to_standard_normal(z_score):
    """
    Compare a Z-score to the standard normal distribution.

    Args:
        z_score (float): The Z-score to compare.

    Returns:
         float: The percentile of the z-score in the standard normal distribution.
    """
    percentile = norm.cdf(z_score)
    return percentile
if __name__ == "__main__":
    data_set = [70, 85, 90, 60, 75, 80, 95]
    data_point = 88
    z_score = calculate_z_score(data_point, data_set)
    percentile = compare_to_standard_normal(z_score)
    print(f"Data point: {data_point}")
    print(f"Data set: {data_set}")
    print(f"Z-score: {z_score:.2f}")
    print(f"Percentile in standard normal distribution: {percentile:.4f}")
    if abs(z_score) > 2:
      print("The data point is considered an outlier.")
    else:
      print("The data point is not considered an outlier.")
```

15.Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np #code
import scipy.stats as stats
n1 = 50
x1 = 75
s1 = 10
n2 = 60
x2 = 80
s2 = 12
D = 0
alpha = 0.05
z_score = ((x1 - x2) - D) / np.sqrt((s1**2 / n1) + (s2**2 / n2))
print('Z-Score:', np.abs(z_score))
z_critical = stats.norm.ppf(1 - alpha/2)
print('Critical Z-Score:',z_critical)
if np.abs(z_score) > z_critical:
    print("""Reject the null hypothesis.
There is a significant difference b/w the online and offline classes.""")
else:
    print("""Fail to reject the null hypothesis.
There is not evidence to suggest a significant difference b/w the online and offline classes.""")
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))
print('P-Value :',p_value)
if p_value < alpha:
    print("""Reject the null hypothesis.
There is a significant difference between the online and offline classes.""")
else:
    print("""Fail to reject the null hypothesis.
There is not evidence to suggest significant difference b/w the online and offline classes.""")
```

16.Create a confidence interval for a dataset using Python and interpret the result.

1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import scipy.stats as stats #code
import numpy as np
def perform_z_test(sample_mean, sample_std_dev, sample_size, population_mean, alpha=0.05):
    """
    Performs a one-sample Z-test to compare a sample mean to a known population mean.

    Args:
        sample_mean: The mean of the sample data.
        sample_std_dev: The standard deviation of the sample data.
        sample_size: The number of observations in the sample.
        population_mean: The known population mean to compare against.
        alpha: The significance level (default: 0.05).

    Returns:
        A dictionary containing the z-statistic, p-value, and interpretation.
    """
    z_statistic = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))
    p_value = 2 * stats.norm.cdf(-abs(z_statistic))
    if p_value < alpha:
        interpretation = "Reject the null hypothesis: The sample mean is significantly different from the population mean."
    else:
        interpretation = "Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean."

    return {
        "z_statistic": z_statistic,
        "p_value": p_value,
        "interpretation": interpretation
    }
if __name__ == "__main__":
    sample_data = [10, 12, 11, 13, 14, 11, 10, 12, 13, 12]
    sample_mean = np.mean(sample_data)
    sample_std_dev = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)
    population_mean = 11.5
    test_results = perform_z_test(sample_mean, sample_std_dev, sample_size, population_mean)
    print("Z-Statistic:", test_results["z_statistic"])
    print("P-value:", test_results["p_value"])
    print("Interpretation:", test_results["interpretation"])
```

3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np #code
from scipy.stats import norm
def one_sample_z_test(sample_data, pop_mean, pop_std_dev, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test.

    Args:
        sample_data (list or numpy.ndarray): The sample data.
        pop_mean (float): The population mean.
        pop_std_dev (float): The population standard deviation.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): Defines the alternative hypothesis.
            Must be one of {'two-sided', 'left-tailed', 'right-tailed'}.
            Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - pop_mean) / (pop_std_dev / np.sqrt(n))
    if alternative == 'two-sided':
        p_value = 2 * norm.cdf(-np.abs(z_statistic))
    elif alternative == 'left-tailed':
        p_value = norm.cdf(z_statistic)
    elif alternative == 'right-tailed':
        p_value = 1 - norm.cdf(z_statistic)
    else:
        raise ValueError("alternative must be one of {'two-sided', 'left-tailed', 'right-tailed'}")
    return z_statistic, p_value
if __name__ == '__main__':
    sample = [85, 90, 78, 92, 88, 84, 91, 80, 87, 82]
    population_mean = 83
    population_std_dev = 5
    alpha_level = 0.05
    z_stat, p_val = one_sample_z_test(sample, population_mean, population_std_dev, alpha_level, alternative='two-sided')
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_val:.4f}")
    if p_val < alpha_level:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
```

5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np #code
import matplotlib.pyplot as plt
from scipy.stats import norm
def visualize_errors(mu0, mu1, sigma, alpha):
    """
    Calculates and visualizes Type I and Type II errors in a hypothesis test.

    Args:
        mu0: Mean of the null hypothesis distribution.
        mu1: Mean of the alternative hypothesis distribution.
        sigma: Standard deviation of both distributions.
        alpha: Significance level (Type I error rate).
    """
    type_1_error = alpha
    z_alpha = norm.ppf(1 - alpha)
    x_critical_value = mu0 + z_alpha * sigma
    beta = norm.cdf(x_critical_value, mu1, sigma)
    power = 1 - beta
    x = np.linspace(mu0 - 3 * sigma, mu0 + 3 * sigma, 100)
    x_alt = np.linspace(mu1 - 3 * sigma, mu1 + 3 * sigma, 100)
    plt.figure(figsize=(8, 5))
    plt.plot(x, norm.pdf(x, mu0, sigma), label='Null Hypothesis (H0)', color='blue')
    plt.plot(x_alt, norm.pdf(x_alt, mu1, sigma), label='Alternative Hypothesis (H1)', color='green')
    plt.axvline(x=x_critical_value, color='red', linestyle='--', label=f'Critical Value (α={alpha})')
    plt.fill_between(x[x > x_critical_value], norm.pdf(x[x > x_critical_value], mu0, sigma), color='red', alpha=0.3, label="Type I Error (False Positive)")
    plt.fill_between(x_alt[x_alt <= x_critical_value], norm.pdf(x_alt[x_alt <= x_critical_value], mu1, sigma), color='blue', alpha=0.3, label="Type II Error (False Negative)")
    plt.xlabel("Sample Mean (x)")
    plt.ylabel("Probability Density")
    plt.title("Visualization of Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()
    print(f"Type I Error (α): {type_1_error:.3f}")
    print(f"Type II Error (β): {beta:.3f}")
    print(f"Power: {power:.3f}")
if __name__ == "__main__":
    mu0 = 10
    mu1 = 12
    sigma = 2
    alpha = 0.05
    visualize_errors(mu0, mu1, sigma, alpha)
```

9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np #code
import scipy.stats
def mean_confidence_interval(data, confidence=0.95):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    data: A list or numpy array of sample data.
    confidence: The desired confidence level (e.g., 0.95 for 95% confidence).

  Returns:
    A tuple containing the sample mean, lower bound of the confidence interval, and upper bound of the confidence interval.
  """
  a = 1.0 * np.array(data)
  n = len(a)
  mean, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return mean, mean - h, mean + h
data = [23, 25, 28, 30, 32, 35, 37, 39]
mean, lower_bound, upper_bound = mean_confidence_interval(data, 0.95)
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
def bayes_theorem(prior_prob, likelihood, evidence_prob): #code
    """
    Calculates the posterior probability using Bayes' Theorem.

    Args:
        prior_prob: The probability of the hypothesis before observing the evidence.
        likelihood: The probability of observing the evidence given the hypothesis.
        evidence_prob: The probability of the evidence.

    Returns:
        The posterior probability (probability of the hypothesis given the evidence).
    """
    posterior_prob = (prior_prob * likelihood) / evidence_prob
    return posterior_prob
prior_prob_disease = 0.01
likelihood_positive_given_disease = 0.95
evidence_prob_positive = 0.10
posterior_prob_disease_given_positive_test = bayes_theorem(
    prior_prob_disease, likelihood_positive_given_disease, evidence_prob_positive
)
print(
    f"The probability of having the disease given a positive test result is: {posterior_prob_disease_given_positive_test:.2%}"
)
```
